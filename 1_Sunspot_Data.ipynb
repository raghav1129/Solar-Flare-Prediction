```python
#Code notebook for converting data into csv format for sunspots 
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/usaf_solar-region-reports_2016.txt', header = None)
```

```python
row_list = []
for i in range(len(df)):
  a = df.iloc[i,0]
  dict1 = {}
  word = [char for char in a]

  Code = {'Code' : ''.join(word[0:2])}
  dict1.update(Code)

  Year = {'Year' : ''.join(word[2:4])}
  dict1.update(Year)

  Month = {'Month' : ''.join(word[4:6])}
  dict1.update(Month)

  Day = {'Day' : ''.join(word[6:8])}
  dict1.update(Day)

  Hours = {'Hours' : ''.join(word[9:11])}
  dict1.update(Hours)

  Minutes = {'Minutes' : ''.join(word[11:13])}
  dict1.update(Minutes)

  Latitude = {'Latitude' : ''.join(word[14:18])}
  dict1.update(Latitude)

  Meridian_Distance = {'Meridian Distance' : ''.join(word[18:20])}
  dict1.update(Meridian_Distance)

  Wilson_Magnetic_Classification = {'Wilson Magnetic Classification' : ''.join(word[20:24])}
  dict1.update(Wilson_Magnetic_Classification)

  Max_Magnetic_Field_Strength = {'Max Magnetic Field Strength' : ''.join(word[25:26])}
  dict1.update(Max_Magnetic_Field_Strength)

  Mt_Wilson_Spot_GN = {'Mt. Wilson Spot GN (MWIL)' : ''.join(word[27:32])}
  dict1.update(Mt_Wilson_Spot_GN)

  MWIL_Subscript = {'MWIL Subscript' : ''.join(word[32:33])}
  dict1.update(MWIL_Subscript)

  NOAA_Number = {'NOAA Number' : ''.join(word[33:38])}
  dict1.update(NOAA_Number)

  NOAA_Number_sbscript = {'NOAA Number sbscript' : ''.join(word[38:39])}
  dict1.update(NOAA_Number_sbscript)

  Zurich_Class = {'Zurich Class' : ''.join(word[39:40])}
  dict1.update(Zurich_Class)

  penumbra_Class = {'penumbra Class' : ''.join(word[40:41])}
  dict1.update(penumbra_Class)

  Compactness_Class = {'Compactness Class' : ''.join(word[41:42])}
  dict1.update(Compactness_Class)

  Number_of_Spots = {'Number of Spots' : ''.join(word[43:45])}
  dict1.update(Number_of_Spots)

  Longitudanal_extent = {'Longitudanal extent' : ''.join(word[46:48])}
  dict1.update(Longitudanal_extent)

  Area = {'Area' : ''.join(word[48:52])}
  dict1.update(Area)

  Individual_Central_Meridian_Passage_year = {'Individual Central Meridian Passage year' : ''.join(word[53:55])}
  dict1.update(Individual_Central_Meridian_Passage_year)

  Individual_Central_Meridian_Passage_month = {'Individual Central Meridian Passage month' : ''.join(word[55:57])}
  dict1.update(Individual_Central_Meridian_Passage_month)

  Individual_Central_Meridian_Passage_day = {'Individual Central Meridian Passage day' : ''.join(word[57:61])}
  dict1.update(Individual_Central_Meridian_Passage_day)

  Regional_Central_Meridian_Passage_year = {'Regional Central Meridian Passage year' : ''.join(word[62:64])}
  dict1.update(Regional_Central_Meridian_Passage_year)

  Regional_Central_Meridian_Passage_month = {'Regional Central Meridian Passage month' : ''.join(word[64:66])}
  dict1.update(Regional_Central_Meridian_Passage_month)

  Regional_Central_Meridian_Passage_day = {'Regional Central Meridian Passage day' : ''.join(word[66:70])}
  dict1.update(Regional_Central_Meridian_Passage_day)

  Station_serial_number = {'Station serial number' : ''.join(word[71:74])}
  dict1.update(Station_serial_number)

  Quality = {'Quality' : ''.join(word[75:76])}
  dict1.update(Quality)

  Station = {'Station' : ''.join(word[76:80])}
  dict1.update(Station)



  row_list.a(dict1)
      
      
```
